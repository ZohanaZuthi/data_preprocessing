Feature scaling is the process of normalizing or standardizing the range of independent variables (features) in a dataset. It ensures that all features contribute equally to machine learning models, preventing some features from dominating due to their larger magnitudes.

Normalization and Standardization

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset=pd.read_csv("StudentPerformanceFactors.csv")
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit(dataset[["Attendance"]])
dataset["Attendance_ss"]=pd.DataFrame(ss.transform(dataset[["Attendance"]]),columns=["x"])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.displot(dataset["Attendance_ss"])

plt.subplot(1,2,2)
plt.title("After")
sns.displot(dataset["Attendance"])
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
mn.fit_transform(dataset[["Hours_Studied"]])
dataset["Hours_Studied_mn"]=mn.transform(dataset[["Hours_Studied"]])
```

```python
dataset.head()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.displot(dataset["Hours_Studied_mn"])

plt.subplot(1,2,2)
plt.title("After")
sns.displot(dataset["Hours_Studied"])
plt.tight_layout()
plt.show()
```
